```python
# 載入基礎套件與代稱
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mean_absolute_error(y, yp):
    """
    計算 MAE
    Args:
        - y: 實際值
        - yp: 預測值
    Return:
        - mae: MAE
    """
    mae = MAE = sum(abs(y - yp)) / len(y)
    return mae

# 定義 mean_squared_error 這個函數, 計算並傳回 MSE
def mean_squared_error(z,zp):
    """
    計算 MSE
    Args:
        - z: 實際值
        - zp: 預測值
    Return:
        - mse: MSE
    """
    mse = MSE = sum((z-zp)**2) / len(z)
    return mse
```

```python
w = 3
b = 0.5
x_lin = np.linspace(0, 100, 101)
y = (x_lin + np.random.randn(101) * 5) * w + b
z = (x_lin + np.random.randn(101) * 5) * w + b

plt.plot(x_lin, z, 'b.', label = 'data points')
plt.title("Assume we have data points")
plt.legend(loc = 2)
plt.show()
```

```python
y_hat = x_lin * w + b
z_hat = x_lin * w + b
plt.plot(x_lin, z, 'b.', label = 'data')
plt.plot(x_lin, z_hat, 'r-', label = 'prediction')
plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()
```

```python
MSE = mean_squared_error(z, z_hat)
MAE = mean_absolute_error(y, y_hat)
print("The Mean squared error is %.3f" % (MSE))
print("The Mean absolute error is %.3f" % (MAE))
```
